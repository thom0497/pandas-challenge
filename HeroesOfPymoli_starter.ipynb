### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
user_data = purchase_data['SN'].value_counts()
print("Total Number of Players")
user_data.count()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = purchase_data["Item Name"].value_counts()
item_count_total = item_count.count()
print("Total Number of Items")
print(item_count_total)


avg_purchase = purchase_data.mean()
print("Average Purchase")
avg_purchase = avg_purchase.astype(float).map("${:,.2f}".format)
print(avg_purchase["Purchase ID"])

total_purchase = purchase_data.count()
print("Total Purchases")
print(total_purchase["Purchase ID"])

total_revenue = purchase_data.sum()
print("Total Revenue")
print(total_revenue["Purchase ID"])
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group = purchase_data.groupby("Gender")

count_by_gender_group = gender_group["SN"].nunique()

gender_pct_breakdown = round(count_by_gender_group / user_data.count() * 100, 2)

gender_demographics = pd.DataFrame({
"Total Count" : count_by_gender_group,
"Percentage of Players" : gender_pct_breakdown.astype(float).map("{:,.2f}%".format)
})
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
analysis_data = purchase_data
gender_group = analysis_data.groupby("Gender")
gender_age_data = gender_group["Age"].mean()
gender_count_data = gender_group["Purchase ID"].count()
gender_purchase_data = gender_group["Purchase ID"].mean()
gender_total_purchase_data = gender_group["Purchase ID"].sum()
gender_price_data = gender_group["Price"].mean()


gender_purchases = pd.DataFrame({
"Average Age" : gender_age_data.astype(float).map("{:,.2f}".format),
"Total Orders" : gender_count_data,
"Gender Purchase" : gender_purchase_data.astype(float).map("${:,.2f}".format),
"Total Purchase Value" : gender_total_purchase_data.astype(float).map("${:,.2f}".format),
"Average Price" : gender_price_data.astype(float).map("${:,.2f}".format)
})

gender_purchases
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_demo = purchase_data

bins = [0, 9, 14, 19, 24, 29, 34, 39, 44, 49]
group_lables = ["0-9", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45-49"]

pd.cut(age_demo["Age"], bins, labels=group_lables)
age_demo["Age Group"] = pd.cut(age_demo["Age"], bins, labels=group_lables)
age_demo.head()

age_group = age_demo.groupby("Age Group")
avg_purchase_age = age_group["Price"].mean()
total_purchase_age = age_group["Purchase ID"].sum()
purchase_count_age = age_group["Purchase ID"].count()
avg_purchase_total_age  = age_group["Purchase ID"].mean()

age_purchased = pd.DataFrame({
"Total Purchases Percent" : round(purchase_count_age/ user_data.count() * 100, 2).astype(float).map("{:,.2f}%".format),
"Total Purchases" : purchase_count_age
})
age_purchased
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_group = age_demo.groupby("Age Group")
avg_purchase_age = age_group["Price"].mean()
total_purchase_age = age_group["Purchase ID"].sum()
purchase_count_age = age_group["Purchase ID"].count()
avg_purchase_total_age  = age_group["Purchase ID"].mean()

age_purchased = pd.DataFrame({
"Total Purchases" : purchase_count_age,
"Average Purchase Price" : avg_purchase_age.astype(float).map("${:,.2f}".format),
"Average Purchase Total" : avg_purchase_total_age.astype(float).map("${:,.2f}".format),
"Total Purchase Value" : total_purchase_age.astype(float).map("${:,.2f}".format)
})
age_purchased
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_group = purchase_data.groupby("SN")
top_spenders_purchase_count = top_spenders_group["Purchase ID"].count()
top_spenders_avg_purchase_price = round(top_spenders_group["Price"].mean(), 2)
top_spenders_avg_total_purchase_value = round(top_spenders_group["Price"].sum(), 2)
top_spenders_df = pd.DataFrame({"Purchase Count" : top_spenders_purchase_count,
                               "Average Purchase Price" : top_spenders_avg_purchase_price.astype(float).map("${:,.2f}".format),
                               "Total Purchase Value" : top_spenders_avg_total_purchase_value.astype(float).map("${:,.2f}".format)
                               })
top_spenders_df.sort_values(by=["Total Purchase Value"], ascending = False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_titles_group = purchase_data.groupby("Item ID")

top_titles_purchase_count = top_titles_group["Purchase ID"].count()
top_titles_avg_purchase_price = round(top_titles_group["Price"].mean(), 2)
top_titles_avg_total_purchase_value = round(top_titles_group["Price"].sum(), 2)
top_titles_name = purchase_data["Item Name"]


top_titles_df = pd.DataFrame({
                              "Title" : top_titles_name,
                            "Purchase Count" : top_titles_purchase_count,
                               "Average Purchase Price" : top_titles_avg_purchase_price.astype(float).map("${:,.2f}".format),
                               "Total Purchase Value" : top_titles_avg_total_purchase_value.astype(float).map("${:,.2f}".format)
})
top_titles_df.head()
```
